```python
n = int(input("Nhập số lượng phần tử: "))
arr = list(map(int, input("Nhập dãy số: ").split()))

tong_chan = 0
tong_le = 0

for x in arr:
    if x % 2 == 0:
       tong_chan += x
    else:
       tong_le += x

print("Tổng số chẵn =", tong_chan)
print("Tổng số lẻ =", tong_le)
```
